### **Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Histograma

data = [12, 15, 14, 16, 18, 20, 15, 14, 17, 19, 18]

plt.hist(data, bins=25, color='yellow')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de mis datos')
plt.show()
```

```python
# 2. Diagrama circular

sizes = [20, 33, 10, 40]
labels = ["A", "B", "C", "D"]

plt.pie(sizes, labels= labels, autopct= "%1.2f%%")
plt.show()
```

```python
# 3. Mapa de calor

mat = np.random.rand(7, 5)
plt.imshow(mat, cmap="hot")
plt.colorbar()
plt.show()
```

```python
#4. Diagrama de caja

data = [np.random.normal(0, 1, 100)]
plt.boxplot(data)
plt.show()
```

```python
# 5. Grafico de barras agrupadas

x = np.arange(3)  # 3 categoras
width = 0.2       # ancho de cada barra

grupo1 = [30, 20, 40]
grupo2 = [40, 10, 10]
grupo3 = [60, 50, 20]
grupo4 = [20, 30, 50]

plt.bar(x - 1.5*width, grupo1, width=width, label='Grupo 1')
plt.bar(x - 0.5*width, grupo2, width=width, label='Grupo 2')
plt.bar(x + 0.5*width, grupo3, width=width, label='Grupo 3')
plt.bar(x + 1.5*width, grupo4, width=width, label='Grupo 4')
plt.xticks(x, ['A', 'B', 'C'])  # nombres de categoras
plt.show()
```

### **Seaborn**

```python
import seaborn as sns
import pandas as pd

# 1. Dispersion simple
df = sns.load_dataset("tips")
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    color="black")
plt.show()
```

```python
# 2. Grafico 3D

x3 = np.random.rand(50)
y3 = np.random.rand(50)
z3 = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.scatter(x3, y3, z3, color="green")

ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
ax.set_zlabel("Eje Z")

plt.show()
```

```python
# 3. Grafico lineplot

df1 = pd.DataFrame({"size": [2, 4, 6, 8, 10, 12], "tip": [3, 8, 1, 5, 11, 9]})
sns.lineplot(data=df1, x="size", y="tip", color="gold")
plt.show()
```
